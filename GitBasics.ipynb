# Git Basics

# What is GIT?

Git allows users to track code changes and manage their project using simple commands.

Git is the world's most popular decentralized version control system. 


# What is Repository?

The main component of Git is a Repository used to contain a project.

A repository can be stored locally or on a website, such as GitHub.

Git allows users to store several different repositories and track each one independently.

# Git Work-Flow Diagram

Working directory:-> Modified files, but untracked and not yet ready for commit.

Staging directory:-> Adding modified files to the staging environment means they are ready for commit.

Committed:-> Snapshots of files from the staging area saved in the commit history.

# Staging

When you want Git to track changes you've made to a certain file, you must add it to the staging area. 

Git recognizes when you modify a file but doesn't track it unless you stage it. 

The staging area represents a layer of security, allowing you to review the changes before committing them.

```python
# You can check which files Git tracks by running:
git status
```

```python
# To add a file to the staging area:
git add 'file name'

# To add all the files to the staging area:
git add .
```

# Commits

A commit represents a check-point for your work, a snapshot of your code at a particular point in time. 

Adding files to the staging area means that they are ready to be committed.

```python
# To check if you have any files ready to be committed:
git status
```

```python
# To commit files in staging area:
git commit -m "Message about the commit"
```

# Reverting

```python
# If we have made mistakes during your project development and want to revert a commit for any reason,
# git revert allows us to do so.
git revert "commit_ID"

# we can find commit_id using:
git log
```
